## Setup wifi

```
sudo raspi-config

sudo nano /etc/wpa_supplicant/wpa_supplicant.conf

ctrl_interface=DIR=/var/run/wpa_supplicant GROUP=netdev
update_config=1
country=GB

network={
 ssid="sciento_2.4Ghz"
 scan_ssid=1
 psk="password"
 key_mgmt=WPA-PSK
}


network={
 ssid="Le Terrier_5"
 scan_ssid=1
 psk="password"
 key_mgmt=WPA-PSK
}

network={
 ssid="AMERICA"
 scan_ssid=1
 psk="password"
 key_mgmt=WPA-PSK
}

```
